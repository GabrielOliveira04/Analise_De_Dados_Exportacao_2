# Importando a Base


```python
import pandas as pd

tabela = pd.read_csv('exportacao_Afghanistan.csv')

display(tabela)
```

# Informações Gerais

```python
tabela.info
display(tabela.describe())
```

# Como foi a evolução das exportações para a Afgnistão ao longo dos anos?


```python
tabela_ano = tabela.groupby('Year').sum()
tabela_ano = tabela_ano[['US$ FOB']]
def formartar(valor):
    return f"${valor:,.2f}"

tabela_ano["US$ FOB"] = tabela_ano["US$ FOB"].map(formatar)
display(tabela_ano)
```

# Quais os produtos mais exportados ao longo de todo o período?


```python
tabela_produto = tabela.groupby(by=['SH4 Description','SH2 Description']).sum()
tabela_produto = tabela_produto[['US$ FOB']]
tabela_produto = tabela_produto.sort_values('US$ FOB',ascending = False)

tabela_produto["US$ FOB"] = tabela_produto["US$ FOB"].map(formatar)
display(tabela_produto)
```

# Qual foi o ano que mais vendeu e a media?

```python
tabela_ano1 = tabela.groupby(by=['Year','SH2 Description']).sum()
tabela_ano1 = tabela_ano1[['US$ FOB']]
tabela_ano1 = tabela_ano1.sort_values('US$ FOB',ascending = False)
tabela_ano1['US$ FOB'] = tabela_ano1['US$ FOB'].map(formatar)
display(tabela_ano1)
```
