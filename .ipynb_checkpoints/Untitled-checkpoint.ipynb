# Como foi a evolução das exportações para a Afgnistão ao longo dos anos?


```python
tabela_ano = tabela.groupby('Year').sum()
tabela_ano = tabela_ano[['US$ FOB']]
def formartar(valor):
    return f"${valor:,.2f}"

tabela_ano["US$ FOB"] = tabela_ano["US$ FOB"].map(formatar)
display(tabela_ano)
```
